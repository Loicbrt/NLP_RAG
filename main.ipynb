```python
prompt_list = ["Décrit moi azumarill","Quelle est le champion de la 6 ème arène dans pokémon diamand et perle ?", " Qui est le personnage principal ?", "Quelles sont les pokémon légendaire de pokémon diamant et perle ?", " Dans quelle ville commence t'on ?", "Comment obtenir une pokéball ?", "Quels sont les talent de herbizarre ?","Qulles type est fort contre combat ?", "Comment avoir un Coatox ?","Quand est sorti le jeu pokémon diamand et perle ?"]
```

```python
Chen_result_list = []
chunks_list = []
without_rag_result_list = []


for prompt in prompt_list:
    result, chunks = professeur_Chen(prompt,"openai")
    Chen_result_list.append(result)
    chunks_list.append(chunks)

    response = client.responses.create(
        model="gpt-5-nano",
        input=prompt
    )
    without_rag_result_list.append(response.output_text)
```

```python
for i in range(len(prompt_list)):
    print("Réponse de Chen : ")
    print(Chen_result_list[i])

    print("Chunk utilisé : ")
    for chunk in chunks_list[i]:
        print(chunk)
    print("Réponse sans Rag : ") 
    print(without_rag_result_list[i])
```

```python
result_list = []
chunks_list = []

varying_method_result_list = []
varying_method_chunks_list = []

for prompt in prompt_list:
    result, chunks = professeur_Chen(prompt,"openai")
    result_list.append(result)
    chunks_list.append(chunks)

    result, chunks = professeur_Chen(prompt,"openai",vector_name="varying_")
    varying_method_result_list.append(result)
    varying_method_chunks_list.append(chunks)
```

```python
for i in range(len(prompt_list)):
    print("Réponse du rag classique: ")
    print(result_list[i])

    print("Chunk utilisé : ")
    for chunk in chunks_list[i]:
        print(chunk)

    print("Réponse du rag avec chunking manuel: ") 
    print(varying_method_result_list[i])

    print("Chunk utilisé : ")
    for chunk in varying_method_chunks_list[i]:
        print(chunk)
```
